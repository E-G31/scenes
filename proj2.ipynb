```python
df=pd.read_csv('pokemon_data.csv', encoding='ISO-8859-1')
```

```python
df['height_m'].value_counts()
df['weight_kg'].value_counts()
```

```python
ul1=df.height_m.mean()+3*df.height_m.std()
ll1=df.height_m.mean()-3*df.height_m.std()
ul2=df.weight_kg.mean()+3*df.weight_kg.std()
ll2=df.weight_kg.mean()-3*df.weight_kg.std()
```

```python
#outlier DETECTION of height
df[(df.height_m>ul1) | (df.height_m<ll1)]
```

```python
#outlier DETECTION of weight
df[(df.weight_kg>ul2) | (df.weight_kg<ll2)]
```

```python
#ABNORMALLY LARGE OR SMALL
abnormal=df[(df.weight_kg>ul2) | (df.weight_kg<ll2) | (df.height_m>ul1) | (df.height_m<ll1)]
abnormal
#26 such pokemons
```

```python
df[['gen', 'hp']]
```

```python
#outlier DETECTION of hp
ul3=df.hp.mean()+3*df.hp.std()
ll3=df.hp.mean()-3*df.hp.std()
df[(df.hp>=ul3) | (df.hp<=ll3)]
new_df=df[(df.hp<ul3) & (df.hp>ll3)]
new_df
```

```python
avg_before = df.groupby('gen')['hp'].mean()
avg_after = new_df.groupby('gen')['hp'].mean()
avg_before, avg_after
#yes the trend is affected a bit for each generation after the outliers have been removed
```

```python
df=new_df
df=df.copy()
df['z_score']=(df.attack-df.attack.mean())/df.attack.std()
df
```

```python
df[(df['z_score']>3) | (df['z_score']<-3)]
#top outliers
```

```python
average_attack_by_type = df.groupby('primary_type')['attack'].mean()
average_attack_by_type
```

```python
average_hp_by_generation = df.groupby('gen')['hp'].mean()
average_hp_by_generation
```

```python
total = len(df)
without_secondary = df['secondary_type'].isna().sum()
with_secondary = total - without_secondary
with_secondary
```

```python
filt=df['primary_type']=='water'
df.groupby(filt)['secondary_type'].count()# true values match the ans
```

```python
df['total_stats'] = df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']].sum(axis=1)
```

```python
avg_legendary = df[df['is_legendary']==1]['total_stats'].mean()
avg_sublegendary = df[df['is_sublegendary']==1]['total_stats'].mean()
avg_mythical = df[df['is_mythical']==1]['total_stats'].mean()
print(avg_legendary, avg_mythical, avg_sublegendary)
# legendary are more powerful since they have higher average stats followed by mythical and sublegendary(thats all i know  because i dont know about pokemons)
```

```python
df_corr=df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']].dropna().corr()
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
sns.heatmap(df_corr, annot=True)
#attack and hp with 0.56, defence and sp_defence with 0.52, sp_attack and sp_defence with 0.5
```

```python
speed = df['speed']
plt.hist(speed,bins=20,log=True,edgecolor='black', color='skyblue')
plt.xlabel('Speed')
plt.ylabel('Count')
plt.title('Distribution of Pokémon Speeds')
plt.tight_layout()
plt.yscale('linear')
plt.grid(True)
plt.show()
```

```python
avg_stats_by_gen = df.groupby('gen')['total_stats'].mean()
avg_stats_by_gen.plot(kind='bar', color='green', edgecolor='black')
plt.xlabel('Generation')
plt.ylabel('Average Total Base Stats')
plt.title('Average Total Base Stats by Generation')
plt.tight_layout()
plt.show()
```

```python
can=df['gigantamax'].count()
cannot=df['gigantamax'].isna().sum()
plt.pie([can,cannot], labels=['can gigantamax', 'cannot gigantamax'], wedgeprops={'edgecolor':'black'})
plt.title('CHART')
plt.tight_layout()
plt.show()
```
